```python
import pandas as pd  

import numpy as np 

bots = pd.read_csv('bots/activity_botscore.csv', low_memory=False) 

print(bots) 

 

def generate_distribution_sample(data, sample_size, dist_size):  

    sample_means = [] 

    for i in range(dist_size): 

        sample = np.random.choice(a = data, size = sample_size)  

        sample_means.append(np.mean(sample)) 

    return sample_means 

 

sample_size = 10000 

dist_size = 38000 
sample_means = generate_distribution_sample(bots['count'], sample_size, dist_size) 
```

```python

import matplotlib.pyplot as plt  

plt.hist(sample_means, bins=50)  

plt.xlabel('count')  

plt.show() 
```

```python
np.sort(sample_means) 

lb = np.percentile(sample_means, 2.5) 

ub = np.percentile(sample_means, 97.5) 

print("conf interval for bootstrap:", (lb, ub)) 

```
