```python
import numpy as np
import matplotlib.pyplot as plt

# Задаємо діапазон x значень
x1 = np.linspace(-2*np.pi, 0, 100)
x2 = np.linspace(0, np.pi, 100)
x3 = np.linspace(np.pi, 2*np.pi, 100)

# Задаємо відповідні значення функції розподілу F(x)
y1 = np.zeros_like(x1)
y2 = np.sin(x2) / 2
y3 = np.ones_like(x3)

# Побудова графіка
plt.plot(x1, y1, 'b-', label='F(x) = 0, для x <= 0')
plt.plot(x2, y2, 'g-', label='F(x) = sin(x)/2, для 0 < x <= π')
plt.plot(x3, y3, 'r-', label='F(x) = 1, для x > π')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Графік функції розподілу F(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rv_bernoulli = stats.bernoulli(p=0.3)
```

```python
rv_bernoulli.rvs(14)
plt.hist(rv_bernoulli.rvs(14),bins=10)
plt.show()
```

```python
rv_binom = stats.binom(100, p=0.9) 

rv_binom.rvs(8) 

plt.hist(rv_binom.rvs(80),bins=10) 

plt.show() 
```

```python
a = 5 

b = 10 

rv_uniform = stats.uniform(a, b - a) 

rv_uniform.cdf(5.5)
```

```python
X = np.linspace(a - 2, b + 2, 100)  

cdf = rv_uniform.cdf(X) 

plt.plot(X, cdf) 

plt.ylabel('’F(x)') 

plt.xlabel('x') 

plt.ylim([0,1.5]) 

plt.title('’Cumulative distribution function for uniform')  

plt.show() 
```

```python
X = np.linspace(a-2 , b + 2, 1000)  

pdf = rv_uniform.pdf(X)  

plt.plot(X, pdf) 

plt.ylabel('f(x)') 

plt.xlabel('x') 

plt.xlim([4, 12]) 

plt.title('PDF for uniform') 
```

```python
mu = 2 

sigma = 0.5 

rv_norm = stats.norm(loc=mu, scale=sigma) 

rv_norm.rvs(17) 
```

```python
x = np.linspace(0, 4, 100)  

cdf = rv_norm.cdf(x)  

plt.plot(x, cdf)  

plt.ylabel('F(x)') 

plt.xlabel('x') 

plt.show() 
```

```python
x = np.linspace(0,4,100)  

pdf = rv_norm.pdf(x) 

plt.plot(x, pdf)  

plt.ylabel('f(x)') 

plt.xlabel('x') 
```
